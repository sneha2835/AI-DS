# Is there any relationship between passenger age and fare

**Use a scatter plot**

```python
sns.scatterplot(data=t,x='Age',y='Fare',hue='Sex',color='lavender',edgecolor='pink',s=100,marker='*' )
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

# What are the pairwise relationships between age, fare and passenger class?

**Use a pairplot**

```python
# Pass the whole DataFrame t to include the 'Survived' column
sns.pairplot(data=t[['Age', 'Fare', 'Pclass']], palette='coolwarm')
plt.show()
```

# What is the corelation between between numerical features in the dataset?

**Compute a correlation matrix and visualize a HeatMap**

```python
t.dtypes
```

```python
numeric_cols = t.select_dtypes(include=['float64', 'int64'])
cor_mat = numeric_cols.corr()
sns.heatmap(cor_mat, annot=True, cmap="Spectral", fmt='.2f')
plt.show()
```

**Replacing Column data with**

**1.Mean/Median for horsepower, price.** 

**2.Mode for categorical columns (number_of_doors, fuel_type).**

```python
s['horsepower'].fillna(s['horsepower'].median(),inplace=True)
s['price'].fillna(s['price'].median(), inplace=True)
s['number_of_doors'].fillna(s['number_of_doors'].mode()[0], inplace=True)
s['fuel_type'].fillna(s['fuel_type'].mode()[0], inplace=True)
```

# String & Categorical Data Preprocessing 

```python
s.dtypes
```

**Convert categorical data into lowercase and remove special characters (make, 
fuel_type, aspiration, etc.).**

```python
columns_to_clean = ['make', 'fuel_type', 'aspiration', 'number_of_doors', 'body_style', 
                    'drive_wheels', 'engine_location', 'engine_type', 'number_of_cylinders', 
                    'fuel_system']
for column in columns_to_clean:
    s[column] = s[column].str.lower().str.replace(r'[^a-z0-9\s]', '', regex=True)
```

```python
s
```

**Standardize number_of_cylinders: 
Convert words ("four", "six", "eight") to integers (4, 6, 8).**

```python
s['number_of_cylinders'].unique()
```

```python
cylinder_map = {
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'eight': 8,
    'twelve': 12
}

s['number_of_cylinders'] = s['number_of_cylinders'].replace(cylinder_map)
s['number_of_cylinders'].head()
```

**Replace "?" or NaN values with appropriate replacements.**

```python
s.replace('?', np.nan, inplace=True)
numerical_cols = s.select_dtypes(include=['float64', 'int64']).columns
s[numerical_cols] = s[numerical_cols].fillna(s[numerical_cols].median())


categorical_cols = s.select_dtypes(include=['object']).columns
for col in categorical_cols:
    s[col].fillna(s[col].mode()[0], inplace=True)

print(s.head())
```

```python
s['number_of_cylinders']
```

**Standardize fuel_system categories (e.g., mpfi, spdi → Multi-Point Fuel Injection, 
Single-Point Fuel Injection).**

```python
s['fuel_system'].unique()
```

```python
fuel_system_map = {
    'mpfi': 'Multi-Point Fuel Injection',
    'mfi': 'Multi-Point Fuel Injection',
    'spfi': 'Single-Point Fuel Injection',
    'spdi': 'Single-Point Fuel Injection',
    '2bbl': 'Two-Barrel Carburetor',
    '4bbl': 'Four-Barrel Carburetor',
    'idi': 'Indirect Fuel Injection',
    '1bbl': 'One-Barrel Carburetor'
}

# Standardize the 'fuel_system' column using the mapping dictionary
s['fuel_system'] = s['fuel_system'].replace(fuel_system_map)

# Check the updated DataFrame
print(s['fuel_system'].unique()) 
```

# Handling Date or Range Values 

**Extract ranges from normalized_losses and create bins (e.g., Low, Medium, High 
loss categories).**

```python
s['normalized_losses'].unique()
```

```python
ranges_nl=[]
for value in s['normalized_losses']:
    if 95 <= value >=120:
        ranges_nl.append('Low Loss')
    elif 121 <= value >=145:
        ranges_nl.append('Medium Loss')
    else:
        ranges_nl.append('HighLoss')
               
s['normalized_losses']=ranges_nl
s['normalized_losses']
```

# Encoding Categorical Features

**Label Encoding for fuel_type, aspiration, drive_wheels, etc.**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_columns = ['fuel_type', 'aspiration', 'drive_wheels']

label_encoder = LabelEncoder()
for column in categorical_columns:
    s[column] = label_encoder.fit_transform(s[column])

print(s.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.figure(figsize=(8,5))
plt.plot(x, y, marker='o', color='b', label='Line plot')
plt.legend()
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
plt.plot(x, y, 
         marker='o',                 # Circular markers
         linestyle='--',             # Solid line
         color='black',              # Blue color
         label='Line plot',          # Label for the plot (used in legend)
         linewidth=3,                # Thicker line
         markersize=10,               # Larger markers
         markerfacecolor='red',      # Red marker fill
         markeredgecolor='orange',   # Black marker edges
         alpha=1)                  # 70% opacity
                  
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[50000,10000,20333,44839,33980,33990]
plt.plot(x,y,
             color='green',
             linestyle='--',
             marker='*',
             label='Savings Money')
plt.legend()
plt.grid(True)
plt.xlabel('Months')
plt.ylabel('Savings amount')
plt.title("Savings Account Data")
plt.show()
```

```python
x=['A','B','C','D','E']
y=[1,4,2,11,9]
plt.bar(x,y, color='skyblue')
plt.title('Simple Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
x=['A','B','C','D','E']
y=[1,4,2,11,9]
plt.bar(x,y, color=['skyblue','black','green','violet','lavender'])
plt.title('Simple Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
ages=[30,22,23,21,54,43,34,33,54,66]
b=[20,30,40,50,60,70]
plt.hist(ages,b,color='crimson',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
x=['A','B','C','D','E']
y=[1,4,2,11,9]
plt.pie(y,labels=x,startangle=90)
plt.show()
```

```python
PL=['Python','Java','C','C#','C++']
Usage=[26,25,19,11,19]
plt.pie(Usage,labels=Usage,colors=['violet','purple','lavender','indigo','skyblue'],autopct='%1.1f%%')
plt.show()
```

```python
x=['A','B','C','D','E']
y=[1,4,2,11,9]
plt.scatter(x,y, color=['skyblue','black','green','violet','lavender'],s=100,edgecolor='black')
plt.title('Simple Scatter Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

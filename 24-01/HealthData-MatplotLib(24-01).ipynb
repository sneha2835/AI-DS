```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.figure(figsize=(8,5))
plt.plot(x, y, marker='o', color='b', label='Line plot')
plt.legend()
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
plt.plot(x, y, 
         marker='o',                 # Circular markers
         linestyle='--',             # Solid line
         color='black',              # Blue color
         label='Line plot',          # Label for the plot (used in legend)
         linewidth=3,                # Thicker line
         markersize=10,               # Larger markers
         markerfacecolor='red',      # Red marker fill
         markeredgecolor='orange',   # Black marker edges
         alpha=1)                  # 70% opacity
                  
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[50000,10000,20333,44839,33980,33990]
plt.plot(x,y,
             color='green',
             linestyle='--',
             marker='*',
             label='Savings Money')
plt.legend()
plt.grid(True)
plt.xlabel('Months')
plt.ylabel('Savings amount')
plt.title("Savings Account Data")
plt.show()
```

```python
x=['A','B','C','D','E']
y=[1,4,2,11,9]
plt.bar(x,y, color='skyblue')
plt.title('Simple Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
x=['A','B','C','D','E']
y=[1,4,2,11,9]
plt.bar(x,y, color=['skyblue','black','green','violet','lavender'])
plt.title('Simple Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
ages=[30,22,23,21,54,43,34,33,54,66]
b=[20,30,40,50,60,70]
plt.hist(ages,b,color='crimson',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
x=['A','B','C','D','E']
y=[1,4,2,11,9]
plt.pie(y,labels=x,startangle=90)
plt.show()
```

```python
PL=['Python','Java','C','C#','C++']
Usage=[26,25,19,11,19]
plt.pie(Usage,labels=Usage,colors=['violet','purple','lavender','indigo','skyblue'],autopct='%1.1f%%')
plt.show()
```

```python
x=['A','B','C','D','E']
y=[1,4,2,11,9]
plt.scatter(x,y, color=['skyblue','black','green','violet','lavender'],s=100,edgecolor='black')
plt.title('Simple Scatter Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
#sales analysis use line plots to track monthy sales and histograms for order quantity distribution

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales =  [3999,3445,6897,7899,1234,5678,9012,2567,4583,5739,7899,8909]
plt.plot(months,sales,linestyle='--',marker='*',markerfacecolor='blue',color='crimson',markersize=10)
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
#order quantity distribution -histogram 

order_quantities = np.random.randint(1, 20, size=100) 
plt.hist(order_quantities, bins=15, color='orange', edgecolor='black')
plt.title('Order Quantity Distribution')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
#customer demographics:visualize age distribution (histogram)and spending patterns(scatter plot)

ages = np.random.randint(18, 80, size=100) 
plt.hist(ages, bins=15, color='skyblue', edgecolor='black')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#customer spending patterns -scatter plot

spending = np.random.randint(50, 200, size=100) 
plt.scatter(ages, spending, color='green', alpha=0.6)
plt.title('Customer Spending Patterns')
plt.xlabel('Age')
plt.ylabel('Spending ($)')
plt.grid(True)
plt.show()
```

```python
#school performance:compare students' marks across subjects(barchart) and identify outliers(box plot)


subjects = ['Math', 'Science', 'English', 'History', 'Art']
marks =  [92, 85, 76, 88,67]
plt.bar(subjects, marks, color='purple')
plt.title('Students Marks Across Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()

```

```python

#IDENTIFYING OUTLINERS

plt.boxplot(marks, vert=False)  
plt.title('Box Plot for Marks Across Subjects')
plt.xlabel('Marks')
plt.show()
```

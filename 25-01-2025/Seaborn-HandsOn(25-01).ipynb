**Develop a LINEPLOT : Mean temperatures of all years**

```python
g['dt']=pd.to_datetime(g['dt'])
```

```python
g.dtypes
```

```python
g.loc[:,'year']= g['dt'].dt.year
```

```python
g.dtypes
```

```python
mean_temp_year=g.groupby('year')['AverageTemperature'].mean()
mean_temp_year
```

```python
plt.plot( mean_temp_year.index, mean_temp_year.values, marker='o', linestyle='--', color='lightpink', markerfacecolor='blue')
plt.title('Mean Temperature by Year')
plt.xlabel('Year')
plt.ylabel('Mean Temperature')
plt.show()
```

**Develop a BAR CHART : Highest average temperature of top 10 countries**

```python
country_avgtemp= g.groupby('Country')['AverageTemperature'].mean().reset_index()
top_countries =country_avgtemp.sort_values(by= 'AverageTemperature', ascending=False).head(10)
plt.figure(figsize=(12, 8))
plt.bar(top_countries['Country'],top_countries['AverageTemperature'],color='lightpink')
plt.xlabel('Country')
plt.ylabel('Average Temperature')
plt.show()
```

**Develop a HISTOGRAM - distribution of average temperatures across all data**

```python
plt.figure(figsize=(10,10))
plt.hist(g['AverageTemperature'],bins=20, color= 'pink',edgecolor='orange')
plt.xlabel('Average Temperature')
plt.ylabel('Frequency')
plt.show()
```

**Develop a SCATTER PLOT - corelation between temperature and uncertainity in the dataset?**

```python
plt.scatter(g['AverageTemperature'],g['AverageTemperatureUncertainty'],color='lavender',edgecolor='pink',s=100,marker='*' )
plt.xlabel('Temperature')
plt.ylabel('Uncertainity')
plt.show()
```

**Develop a BOX PLOT - Spread of Temperature for a specfic country over a given period**

```python
country = 'Denmark'
year = 2013
data = g[(g['Country'] == country) & (g['year'] == year)]
plt.boxplot(x=data['AverageTemperature'])
plt.title(f'Spread of Temperature in {country} for the Year {year}')
plt.ylabel('Average Temperature (°C)')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.histplot(data=tips, x='total_bill', bins=20, color='skyblue',kde=True)
plt.title("Distribution of Total Bills")
plt.show()
```

```python
sns.kdeplot(data=tips, x='total_bill', color='skyblue',fill=True)
plt.title("Distribution of Total Bills")
plt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill',palette='viridis')
plt.show("Bar Plot of Total Bills")
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue="sex",palette='coolwarm')
plt.title("Box Plot of Total Bills by Day and Gender")
plt.show()
```

```python
sns.scatterplot(data=tips,x='day',y='tip',hue="day",style='time',palette='deep')
plt.title("Scatter Plot of Total Bills vs Tip")
plt.show()
```

```python
sns.lineplot(data=tips,x='size',y='tip',marker='o',color='green')
plt.title("Line party of party size vs tip")
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='husl')
plt.show()
```

```python
numeric_cols = tips.select_dtypes(include=['float64', 'int64'])
cor_mat = numeric_cols.corr()
sns.heatmap(cor_mat, annot=True, cmap="coolwarm", fmt='.2f')
plt.show()
```

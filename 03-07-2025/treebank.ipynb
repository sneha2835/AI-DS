What fraction of JJ words are followed by a noun word in corpus


```python
taggs=[pair[1] for pair in tag_words]
taggs
```

```python
jj_tags=[pair for pair in tag_words if pair[1]=='JJ']
```

```python
#creating a list of JJ tags
jj_nn_tags=[(t,taggs[index+1]) for index,t in enumerate(taggs)
              if t=='JJ' and taggs[index+1]=='NN']
jj_nn_tags
```

```python
print(len(jj_tags))
print(len(jj_nn_tags))
print(len(jj_nn_tags)/len(jj_tags))
```

```python
md_tags=[pair for pair in tag_words if pair[1]=='MD']
```

```python
print(len(md_tags))
print(len(md_vb_tags))
print(len(md_vb_tags)/len(md_tags))
```

Lexicon unigram tagger

```python
uni_tag=nltk.UnigramTagger(train_set)
```

```python
uni_tag.accuracy(test_set)
```

Ruled based (reg ecp based tagger) Tagger

```python
patterns=[
    (r'.*ing$','VBG'),
    (r'.*ed$','VBD'),
    (r'.*es$','VBZ'),
    (r'.*ould$','MD'),
    (r'.*\'s$','NN'),
    (r'.*s$','NNS'),
    (r'^-?[0-9]+(.[0-9]+)?$','CD'),
    (r'.*','NN')
]
```

```python
reg_tag=nltk.RegexpTagger(patterns)
```

```python
reg_tag.evaluate(test_set)
```
